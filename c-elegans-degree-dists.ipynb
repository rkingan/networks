# C. *Elegans* degree distributions

Bar charts showing in-degree and out-degree distributions for the C. *Elegans* connectome. Source data is from files `in-degree-dist.csv` and `out-degree-dist.csv` in dev/comp/c-elegans.

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import csv
```

```python
in_fname = "in-degree-dist.csv"
out_fname = "out-degree-dist.csv"
with open(in_fname, encoding="utf-8") as f:
    reader = csv.reader(f)
    next(reader)  # skip header
    in_degs, in_freqs = zip(*((int(r[0]), int(r[1])) for r in reader))

ni = len(in_degs)
    
with open(out_fname, encoding="utf-8") as f:
    reader = csv.reader(f)
    next(reader)  # skip header
    out_degs, out_freqs = zip(*((int(r[0]), int(r[1])) for r in reader))
    
no = len(out_degs)
```

```python
plt.rcParams.update({"font.size": 6})
fig, axs = plt.subplots(2, 1, figsize=(6, 9), dpi=600)
in_inds = list(range(ni))
axs[0].bar(in_inds, in_freqs, color="black", width=.1)
axs[0].set_xticks(in_inds)
axs[0].set_xticklabels([str(x) for x in in_degs])
axs[0].set_xlabel("Degree", fontsize=8)
axs[0].set_ylabel("Frequency", fontsize=8)

out_inds = list(range(no))
axs[1].bar(out_inds, out_freqs, color="black", width=.1)
axs[1].set_xticks(out_inds)
axs[1].set_xticklabels([str(x) for x in out_degs])
axs[1].set_xlabel("Degree", fontsize=8)
axs[1].set_ylabel("Frequency", fontsize=8)

fig.savefig("c-elegans-degree-dists.jpg")
plt.show()
```
